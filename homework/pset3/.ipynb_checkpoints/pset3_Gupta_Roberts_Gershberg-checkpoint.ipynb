```python
custom_doj_stopwords = ["civil", "rights", "division", "department", "justice",
                        "office", "attorney", "district", "case", "investigation", "assistant",
                       "trial", "assistance", "assist"]
```

```python
## your code defining a text processing function
updated_stopwords = stopwords.words("english") + custom_doj_stopwords
snowball = SnowballStemmer(language='english')
def text_processor(content):
    content_lower = content.lower()
    content_clean = [snowball.stem(word) for word in word_tokenize(content_lower) if word not in updated_stopwords and word.isalpha() and len(word) >= 4]
    return ' '.join(content_clean)   
```

A. Use the `create_dtm` function I provide (alternately, feel free to write your own!) and create a document-term matrix using the preprocessed press releases; make sure metadata contains the following columns: `id`, `compound` sentiment column you added, and the `topics_clean` column

For steps B - D, to receive full credit, write a function `get_topwords` that helps you avoid duplicated code when you find top words for the different subsets of the data. There are different ways to structure it but one way is to feed it subsetted data (so data subsetted to one topic etc.) and for it to get the top words for that subset.


```python
def create_dtm(list_of_strings, metadata):
    vectorizer = CountVectorizer(lowercase = True)
    dtm_sparse = vectorizer.fit_transform(list_of_strings)
    dtm_dense_named = pd.DataFrame(dtm_sparse.todense(), columns=vectorizer.get_feature_names_out())
    # dtm_dense_named = pd.DataFrame(dtm_sparse.todense(), columns=vectorizer.get_feature_names())
    dtm_dense_named_withid = pd.concat([metadata.reset_index(drop=True), dtm_dense_named], axis = 1)
    return dtm_dense_named_withid
```

```python
# your code here
# PART A
# compound is part of corpus & needs to be renamed in metadata
dtm = create_dtm(list_of_strings=doj_subset_wscore["processed_text"], metadata=doj_subset_wscore[['id', 'compound', 'topics_clean']].rename(columns={"compound": "compound_sentiment"}))
dtm
```

```python
# Function for Parts B-D
def find_top_terms(doc_matrix):
    top_10_terms = doc_matrix[doc_matrix.columns[3:]].sum(axis = 0)
    return top_10_terms.sort_values(ascending = False)[:10]
```

```python
# PART B
doc_matrix_95 = dtm[dtm["compound_sentiment"] >= dtm["compound_sentiment"].quantile(0.95)]
top_10_95 = find_top_terms(doc_matrix_95)
top_10_95
```

```python
# PART C
doc_matrix_5 = dtm[dtm["compound_sentiment"] <= dtm["compound_sentiment"].quantile(0.05)]
top_10_5 = find_top_terms(doc_matrix_5)
top_10_5
```

# 3. Extend the analysis from unigrams to bigrams (10 points)

In the previous question, you found top words via a unigram representation of the text. Now, we want to see how those top words change with bigrams (pairs of words)

A. Using the `doj_subset_wscore` data and the `processed_text` column (so the words after stemming/other preprocessing), create a column in the data called `processed_text_bigrams` that combines each consecutive pairs of word into a bigram separated by an underscore. Eg:

"depart reach settlem" would become "depart_reach reach_settlem"

Do this by writing a function `create_bigram_onedoc` that takes in a single `processed_text` string and returns a string with its bigrams structured similarly to above example
 
**Hint**: there are many ways to solve but `zip` may be helpful: https://stackoverflow.com/questions/21303224/iterate-over-all-pairs-of-consecutive-items-in-a-list

B. Print the `id`, `processed_text`, and `processed_text_bigram` columns for press release with id = 16-217

```python
## your code here 
def create_bigram_onedoc(text):
    tokens = word_tokenize(text)
    bigram_text = ""
    for first_word, second_word in zip(tokens, tokens[1:]):
        bigram_text += first_word + "_" + second_word + " "

    return bigram_text   
```

```python
doj_subset_wscore["processed_text_bigrams"] = doj_subset_wscore["processed_text"].apply(create_bigram_onedoc)
doj_subset_wscore[doj_subset_wscore["id"] == "16-217"][["id", "processed_text", "processed_text_bigrams"]]
```

#### The bigram matrix has more columns than the unigram matrix because now the frequency of 2 consecutive words is being recorded rather than just the presence of a single word in the press release. With a bigram matrix, each column represents 2 words that were presented in a certain order in the press release. Furthermore, there are more combinations per word, meaning that there are more columns, since each word is now joined with the word preceding it as well as the word that comes afterwards. So, each word is part of at least 1-2 bigram phrases that make up the columns of the bigram matrix.

```python
# your code here 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf_vectorizer = TfidfVectorizer()
def get_string_similarity(press1, press2):
    press_vectors = tfidf_vectorizer.fit_transform([press1, press2])
    similarity_score = cosine_similarity(press_vectors[0], press_vectors[1])
    return similarity_score[0][0]
```
